```python
import numpy as np
import sys
sys.path.append('../')
from scipy.stats import uniform, norm
from sklearn.linear_model import LinearRegression

def solution():
    x1 = uniform.rvs(size=100,loc=0,scale=1,random_state=1)
    x2 = (0.5*x1)+norm.rvs(size=100,loc=0,scale=1,random_state=1)
    y = 2+(2*x1)+(0.3*x2)+norm.rvs(size=100,loc=0,scale=1,random_state=1)
    mean_y = round(y.mean(),2)

    y_new=y.reshape(-1,1)
    x_new=np.concatenate([x1,x2]).reshape(-1,2)
    lr=LinearRegression()
    lr.fit(x_new,y_new)
    coef=np.round(lr.coef_,1)
    coef=str(coef.tolist())[1:-1]
    intercept=str(np.round(lr.intercept_,1))[1:-1]
    print(mean_y)
    print(str(coef)[1:-1])
    print(str(intercept)[1:-1])
    np.savetxt('output111111.csv',(mean_y,coef,intercept),delimiter=',',fmt='%s',header='')
solution()
```
